```python
# We will drop the rows having missing values in Item Number.
df1 = df1.dropna(axis=0, subset=['Item Number'])
```

# Drop Nulls

```python
# Item Class has null values
pd.unique(df1['Item Class'])
```

```python
# replacing null values with P01
df1 = df1.fillna('P01')
```

```python
# Checking again for null vakues in Item Class
pd.unique(df1['Item Class'])
```

```python
# replace spaces with underscores and lowercase Labels for datasets
df1.rename(columns = lambda x : x.strip().lower().replace(" ","_"),inplace = True)

# confirm changes
df1.head(1)
```

# Filter

```python
# We can drop rows with Item Class as PO1 as they are very less as compared to P01
df1 = df1.query("item_class == 'P01'")
```

```python
# Dataset containing only item_class of P01
df1
```

```python
# Now we can drop item_class column as all the data is for P01
df1 = df1.drop('item_class', axis = 1)
```

```python
df1
```

```python
# Dropping unnecessary columns
unnecessary_columns = ['invoice_number','order_number']
df2 = df1.drop(unnecessary_columns, axis = 1)
df2
```
